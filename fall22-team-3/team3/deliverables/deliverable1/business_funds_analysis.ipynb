```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("funds2.0.csv")
df = df.dropna()
df.reset_index(drop=True)
```

```python
df1 = df.groupby(['Neighborhood']).size().reset_index(name='counts')
df1
```

```python
df1 = df1.sort_values(by=['counts'])
```

```python
neighborhood = df1['Neighborhood'].tolist()
cnt = df1['counts'].tolist()
# plt.bar(neighborhood, cnt, align='center', alpha=0.5)
# y_pos = np.arange(len(neighborhood))
# plt.xticks(y_pos, )
# plt.ylabel('')
# plt.title('')

# plt.show()

plt.figure(figsize=(20,10))
y_pos = np.arange(len(neighborhood))
a = plt.barh(y_pos, cnt, align='center', alpha=0.5)
plt.yticks(y_pos, neighborhood)
plt.xlabel('Small Business')
plt.title('Small business in each area')

plt.tight_layout()
plt.show()
```
