```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("funds2.0.csv")
df = df.dropna()
df.reset_index(drop=True)
```

```python
df1 = df.groupby(['Neighborhood']).size().reset_index(name='counts')
df1
```

```python
list =[4.4, 0.5, 0.2, 31.97, 59.3, 0.2, 1.31, 1.5, 0.5, 0.36, 4.7, 1.24, 0.55, 4.4, 2.8, 0.75, 0.37, 2.59, 26.3, 13.15, 0.15, 0.82, 4.61]
df1['Area'] = pd.DataFrame(list)
```

```python
df1 = df1.sort_values(by=['counts'])
```

```python
df1
```

```python
df1['Density'] = df1['counts']/df1['Area']
```

```python
df1
```

```python
tmp = df1
tmp.set_index('Neighborhood', inplace=True)
```

```python
tmp['Density'].sort_values().plot.pie(autopct='%.1f', counterclock=False, figsize=(10,10))
plt.ylabel('')
plt.title('Density of business', fontsize=24,fontweight='bold')
plt.legend(loc=2, bbox_to_anchor=(1.2,1.0))
```

```python
df2 = df1.sort_values(by=['Density'])
```

```python
neighborhood = df2['Neighborhood'].tolist()
dens = df2['Density'].tolist()

plt.figure(figsize=(20,10))
y_pos = np.arange(len(neighborhood))
a = plt.barh(y_pos, dens, align='center', alpha=0.5)
plt.yticks(y_pos, neighborhood)
plt.xlabel('Small Business')
plt.title('Density of business in each area', fontsize=16,fontweight='bold')

plt.tight_layout()
plt.show()
```
