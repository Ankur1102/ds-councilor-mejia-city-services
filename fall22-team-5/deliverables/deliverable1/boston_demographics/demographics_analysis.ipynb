```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_neighborhoods = pd.read_csv("./boston-neighborhood-data.csv")
```

```python
df_neighborhoods = df_neighborhoods.iloc[ : , :7 ]

df_neighborhoods.columns = df_neighborhoods.iloc[0] # change the headings to be understandable

df_neighborhoods = df_neighborhoods[1:] # removes redundant row of column headings

df_neighborhoods.iloc[ : , 1: ] = df_neighborhoods.iloc[ : , 1: ].astype(int)

df_neighborhoods
```
