# Budget Survey Response Analysis 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")
```

```python
df = pd.read_csv("budget-survey.csv")
df
```

Removing rows with all null values

```python
df = df.dropna(how='all')
df
```

Removing rows with more than 3 null values

```python
df = df.dropna(thresh=len(df.columns)-3)
df
```
