```python
df1_temp = df2[["Neighborhood", "Total RRF Award"]].dropna()
df_count = df1_temp["Total RRF Award"].str.split(r'[$,]').apply(lambda numbers: ''.join(numbers))
#df_count = df_count["Total RRF Award"].str.split(",").apply(lambda numbers: ''.join(numbers))
df1_temp["Total RRF Award"] = df_count.astype('float')
df1_temp.head()
df1_temp.describe()
```

```python
df1_temp = df1_temp.groupby(["Neighborhood"], as_index=False).sum().rename(columns={'Total RRF Award':'Total Award'})
print(df1_temp)
```

```python
import numpy as np
import matplotlib.pyplot as plt

XScale = df1_temp["Neighborhood"]
YScale = df1_temp["Average Award"]

fig = plt.figure(figsize = (15, 10))

plt.bar(XScale, YScale, color = "blue", width = 0.5)
plt.xticks(rotation = 90)

plt.xlabel("Neighborhood")
plt.ylabel("Average Rental Assistance Award")
plt.title("Average Rental Assistance Award by Neighborhood")

plt.show 
```

```python
df_count = df1_temp[["Neighborhood", "Total RRF Award"]].dropna().groupby(["Neighborhood"], as_index=False).sum()

labels = df_count["Neighborhood"]
total_sum = df_count["Total RRF Award"].sum()
df_count["RRF_Prop"] = df_count["Total RRF Award"]/total_sum
sizes = df_count["RRF_Prop"]
#explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots(figsize =(20, 14))

ax1.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=-60)
ax1.set_title("Rental Assistance by Neighborhood", fontweight ='bold')

ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
