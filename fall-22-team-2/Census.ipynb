```python
import pandas as pd
race = pd.read_csv('boston-neighborhood-data.csv')
```

```python
race.head()
```

```python
test = race.iloc[:,:7]
```

```python
test.head()
```

```python
from matplotlib import pyplot as plt
```

```python
len(test.index)
print(test.loc[3, :].values.flatten().tolist())
```

```python
fig,ax = plt.subplots(8, 3, figsize=(15, 25))
plt.tight_layout()
def get_sizes(list):
    sizes = []
    for i in list[2:]:
        sizes.append(int(i)/int(list[1]))
    return sizes

labels = ["White", "Black", "Hispanic/Latino", "AAPI", "Other"]
row = 1
for i in list(range(8)):
    for j in list(range(3)):
        sizes = get_sizes(test.loc[row, :].values.flatten().tolist())
        ax[i][j].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=-60)
        ax[i][j].set_title((test.loc[row, :].values.flatten().tolist())[0])
        if row < 23:
            row += 1
        else:
            break
```
